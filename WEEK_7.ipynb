```python
# Naive Bayes from scratch (Categorical Data)

import pandas as pd
from collections import defaultdict
import math

# ---------------------------
# Dataset 3: Loan Approval
# ---------------------------
data = {
    "Weather": ["Rain", "Rain", "Sunny", "Rain", "Sunny",
                "Sunny", "Rain", "Sunny", "Sunny", "Rain"],
    "RoadCondition": ["Good", "Bad", "Good", "Normal", "Good",
                      "Good", "Normal", "Bad", "Normal", "No"],
    "Traffic": ["Normal", "Heavy", "Normal", "Normal", "Heavy",
                "Normal", "Normal", "Heavy", "Normal", "Normal"],
    "Accidents": ["Yes", "Yes", "No", "Yes", "No",
                  "No", "No", "No", "Yes", "Yes"]
}

df = pd.DataFrame(data)
df
```

```python
total = len(df)

p_yes = len(df[df["Accidents"] == "Yes"]) / total
p_no = len(df[df["Accidents"] == "No"]) / total

print("P(Accidents = Yes) =", p_yes)
print("P(Accidents = No) =", p_no)
```

```python
def conditional_probability(feature, value, target_value):
    target_data = df[df["Accidents"] == target_value]
    return len(target_data[target_data[feature] == value]) / len(target_data)

print("P(Rain | Accidents=Yes) =", conditional_probability("Weather", "Rain", "Yes"))
print("P(Good | Accidents=Yes) =", conditional_probability("RoadCondition", "Good", "Yes"))
print("P(Normal | Accidents=Yes) =", conditional_probability("Traffic", "Normal", "Yes"))

print("P(Rain | Accidents=No) =", conditional_probability("Weather", "Rain", "No"))
print("P(Good | Accidents=No) =", conditional_probability("RoadCondition", "Good", "No"))
print("P(Normal | Accidents=No) =", conditional_probability("Traffic", "Normal", "No"))
```

```python
# Values to predict
weather = "Rain"
road = "Good"
traffic = "Normal"

# Accidents = Yes
p_yes_x = (
    p_yes *
    conditional_probability("Weather", weather, "Yes") *
    conditional_probability("RoadCondition", road, "Yes") *
    conditional_probability("Traffic", traffic, "Yes")
)

# Accidents = No
p_no_x = (
    p_no *
    conditional_probability("Weather", weather, "No") *
    conditional_probability("RoadCondition", road, "No") *
    conditional_probability("Traffic", traffic, "No")
)

print("P(Accidents=Yes | X) =", p_yes_x)
print("P(Accidents=No | X) =", p_no_x)
```

```python
if p_yes_x > p_no_x:
    print("Prediction: Accidents = Yes")
else:
    print("Prediction: Accidents = No")
```
